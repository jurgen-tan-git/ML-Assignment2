```python
# new_feature = X.iloc[:,5] + X.iloc[:,0]
new_df = pd.DataFrame()
new_df["peak_amplitude"] = X.iloc[:,0]
new_df["rise_time"] = X.iloc[:,4]
new_df["pulse_width"] = X.iloc[:,5]

new_df2 = pd.DataFrame()
new_df2["phase_angle"] = X.iloc[:,2]
new_df2["integral"] = X.iloc[:,1]
new_df2["cycle_number"] = X.iloc[:,3]
```

```python
from sklearn.decomposition import KernelPCA
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(new_df)
X_scaled2 = scaler.fit_transform(new_df2)

# Initialize pca with 1 components
kpca = KernelPCA(n_components=1, kernel='cosine', gamma=10)
kpca2 = KernelPCA(n_components=1, kernel='poly', gamma=10)
# pca = PCA(n_components=1)

# Fit the data
X_kpca = kpca.fit_transform(X_scaled)
X_kpca2 = kpca2.fit_transform(X_scaled2)
# X_pca = pca.fit_transform(X_scaled)
```

## Splitting the dataset into Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

## Training CatBoost on the Training Set

```python
from catboost import CatBoostClassifier
classifier = CatBoostClassifier()
classifier.fit(X_train, y_train)
```

## Predicting the Test Data

```python
y_pred = classifier.predict(X_test)
```

```python
pred_data = pd.DataFrame({"Predicted values":y_pred, "Actual values":y_test})
```

```python
pred_data
```

## Confusion Matrix and Accuracy Score

```python
from sklearn.metrics import confusion_matrix, accuracy_score, balanced_accuracy_score

# First row [Predicted Class 0, Actual Class 0, Predicted Class 0, Actual Class 1]
# Second row [Predicted Class 1, Actual Class 0, Predicted Class 1, Actual Class 1]
cm = confusion_matrix(y_test, y_pred)
print(cm)
tn, fp, fn, tp = cm.ravel()
false_negative_count = fn
false_positive_count = fp
print("False Negative:", false_negative_count)
print("False Positive:", false_positive_count)
print("Accuracy score:", accuracy_score(y_test, y_pred))
print("Balanced Accuracy Score:", balanced_accuracy_score(y_test, y_pred))
```

## F1 Score

```python
from sklearn.metrics import f1_score
f1 = f1_score(y_test, y_pred)
print("F1 score:", f1)
```
