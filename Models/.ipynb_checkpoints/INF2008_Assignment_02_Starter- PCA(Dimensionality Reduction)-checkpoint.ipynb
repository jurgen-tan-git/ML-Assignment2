```python
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
# Standardize the data
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)

# Initialize pca with 2 components for visualization
pca = PCA(n_components = 0.9)

# Fit the data
X = pca.fit_transform(df_scaled)
```

```python
print("Shape of the new reduced data:", X.shape)
```

```python
import matplotlib.pyplot as plt
plt.scatter(X[:, 0], X[:, 1])
plt.xlabel('PCA Component 1')
plt.ylabel('PCA Component 2')
plt.show()
```

```python
import numpy as np
explained_variances = pca.explained_variance_ratio_
cumulative_explained_variances = np.cumsum(explained_variances)
plt.plot(cumulative_explained_variances)
plt.xlabel('Number of principal components')
plt.ylabel('Cumulative explained variance')
plt.show()
```
