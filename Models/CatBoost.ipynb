## Importing the libraries

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import json
import os
```

## Importing the dataset

```python
file_dir = os.getcwd() + "\\Data\\df_full_features.csv"
dataset = pd.read_csv(file_dir)
X = dataset.iloc[:, :-1]
y = dataset.iloc[:, -1]
```

```python
X
```

```python
y
```

## Splitting the dataset into Training set and Test set

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

## Training CatBoost on the Training Set

```python
from catboost import CatBoostClassifier
classifier = CatBoostClassifier()
classifier.fit(X_train, y_train)
```

## Predicting the Test Data

```python
y_pred = classifier.predict(X_test)
```

```python
pred_data = pd.DataFrame({"Predicted values":y_pred, "Actual values":y_test})
```

```python
pred_data
```

## Confusion Matrix, Accuracy Score, Balanced Accuracy Score

```python
from sklearn.metrics import confusion_matrix, accuracy_score, balanced_accuracy_score

# First row [Predicted Class 0, Actual Class 0, Predicted Class 0, Actual Class 1]
# Second row [Predicted Class 1, Actual Class 0, Predicted Class 1, Actual Class 1]
cm = confusion_matrix(y_test, y_pred)
print(cm)
tn, fp, fn, tp = cm.ravel()
false_negative_count = fn
false_positive_count = fp
print("False Negative:", false_negative_count)
print("False Positive:", false_positive_count)
print("Accuracy score:", accuracy_score(y_test, y_pred))
print("Balanced Accuracy Score:", balanced_accuracy_score(y_test, y_pred))
```

## F1 Score

```python
from sklearn.metrics import f1_score
f1 = f1_score(y_test, y_pred)
print("F1 score:", f1)
```
